# Dealing with Outliers

```python
# ploting outliers using boxplot
plt.figure(figsize=(8,5))
sns.boxplot(df['Price'])
```

```python
#Finding IQR values
Q1 = np.percentile(df['Price'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['Price'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1

#Printing Q1 value
print(Q1)

#Printing Q3 value
print(Q3)

#printing IQR value
print(IQR)
```

```python
# Above Upper bound
upper = df['Price'] >= (Q3+1.5*IQR)
 
print("Upper bound:",upper)
print(np.where(upper))
 
# Below Lower bound
lower = df['Price'] <= (Q1-1.5*IQR)
print("Lower bound:", lower)
print(np.where(lower))

#Old shape

# Upper bound
upper = np.where(df['Price'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['Price'] <= (Q1-1.5*IQR))
```

# Bivariate Analysis

Multivariate analysis is the analysis of more than one variable.

```python
#barplot (categorical-numerical)
plt.figure(figsize=(8,5))
plt.title("Prices of Airline")
sns.barplot(data=df, x='Airline',y='Price',color='red')
```

```python
#Average price by source_city
fig, ax = plt.subplots(figsize=(15, 6))
df.groupby(['Source'])['Price'].mean().sort_values(ascending=False).plot(kind = 'bar', width= 0.8)
ax.set(title = "Average Price By Source City",
       xlabel = "Source",
       ylabel = "Average Price")
plt.setp(ax.get_xticklabels(), rotation = 0)
plt.box(False)
plt.show()
```

```python
#Average price by departure time
fig, ax = plt.subplots(figsize=(15, 6))
df.groupby(['Dep_Time'])['Price'].mean().sort_values(ascending=False).plot(kind = 'bar', width= 0.8,color='g')
ax.set(title = "Average Price By Departure Time",
       xlabel = "Departure Time",
       ylabel = "Average Price")
plt.setp(ax.get_xticklabels(), rotation = 0)
plt.box(False)
plt.show()
```

```python
#Average price by days left for flight
fig, ax = plt.subplots(figsize=(15, 6))
df.groupby(['Additional_Info'])['Price'].mean().plot(kind = 'bar', width= 0.8,color='yellow')
ax.set(title = "Average Price ",
       xlabel = "Days Left For Flight",
       ylabel = "Average Price")
plt.setp(ax.get_xticklabels(), rotation = 0)
plt.box(False)
plt.show()
```

# Multivariate analysis

```python
plt.figure(dpi = 200)
sns.pairplot(df,hue = 'Total_Stops',palette = 'plasma')
plt.legend(['Non Diabetic','Diabetic'])
plt.show()
```

```python
plt.figure(dpi = 120,figsize= (5,4))
mask = np.triu(np.ones_like(df.corr(),dtype = bool))
sns.heatmap(df.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
plt.yticks(rotation = 0)
plt.xticks(rotation = 90)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot of Price with Other Variables ==> \n")
for i in  df.columns:
    if i != 'Price'  and i != 'Airline' and i != 'Source' and i != 'Destination' and i != 'flight' and i != 'Dep_Time' and i != 'Arrival_Time' and i != 'Date_of_Journey' and i != 'Route' and i != 'Duration' and i != 'Total_Stops' and i != 'Additional_Info':
        print(f"Correlation between price and {i} ==> ",df.corr().loc['Price'][i])
        sns.jointplot(x='Price',y=i,data=df,kind = 'reg',color = 'red')
        plt.show()
```
